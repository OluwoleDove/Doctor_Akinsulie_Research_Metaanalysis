# **IMPORT MODULES**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Mounting Google Drive (if required, skip if already mounted)
from google.colab import drive
drive.mount('/content/drive')
```

# **READ DATA**

```python
# File paths
file_paths = {
    "c1391bl": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1391bl.xls",
    "c1392bl": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1392bl.xls",
    "c1393bl": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1393bl.xls",
    "c1394bl": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1394bl.xls",
    "c1395bl": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1395bl.xls",
    "c1771": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1771.xls",
    "c1774": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1774.xls",
    "c1775": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1775.xls",
    "c1778": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1778.xls",
    "c1779": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1779.xls",
    "c1798": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1798.xls"
}

# Function to read data from Excel files
def read_data(file_paths):
    data = {}
    for file_name, file_path in file_paths.items():
        df = pd.read_excel(file_path)
        data[file_name] = df
    return data

# Separate file paths into two categories
post_tick_files = {key: file_paths[key] for key in ['c1391bl', 'c1392bl', 'c1393bl']}
post_inoculation_files = {key: file_paths[key] for key in ['c1394bl', 'c1395bl', 'c1771', 'c1774', 'c1775', 'c1778', 'c1779', 'c1798']}

# Read data for both categories
post_tick_data = read_data(post_tick_files)
post_inoculation_data = read_data(post_inoculation_files)
```

# **PLOT CHARTS**

```python
# Set Seaborn style and context
sns.set(style="darkgrid", context="notebook", palette="muted")

# Plotting function
def plot_data(data, x_col, y_col, title, xlabel, ylabel):
    plt.figure(figsize=(10, 6))
    for file, df in data.items():
        sns.lineplot(data=df, x=x_col, y=y_col, label=file)
    plt.title(title, fontsize=16, fontweight='bold')
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    plt.legend(title='Files', fontsize=12)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    sns.despine(trim=True)
    plt.tight_layout()
    plt.show()

# Plot for post-application of ticks
plot_data(post_tick_data, 'Days post-application of ticks', 'PCV',
          'PCV vs Days post-application of ticks', 'Days post-application of ticks', 'PCV')
plot_data(post_tick_data, 'Days post-application of ticks', 'PPE',
          'PPE vs Days post-application of ticks', 'Days post-application of ticks', 'PPE')

# Plot for post-inoculation of blood stabilate
plot_data(post_inoculation_data, 'Days post-inoculation of blood stabilate', 'PCV',
          'PCV vs Days post-inoculation of blood stabilate', 'Days post-inoculation of blood stabilate', 'PCV')
plot_data(post_inoculation_data, 'Days post-inoculation of blood stabilate', 'PPE',
          'PPE vs Days post-inoculation of blood stabilate', 'Days post-inoculation of blood stabilate', 'PPE')
```

```python
# Set Seaborn style and context
sns.set(style="darkgrid", context="notebook", palette="bright")

# Plotting function
def plot_data(data, x_col, y_col, title, xlabel, ylabel):
    plt.figure(figsize=(10, 6))
    for file, df in data.items():
        sns.lineplot(data=df, x=x_col, y=y_col, label=file, marker='o')
    plt.title(title, fontsize=16, fontweight='bold')
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    plt.legend(title='Files', fontsize=12, loc='upper right')
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    sns.despine(trim=True)
    plt.tight_layout()
    plt.show()

# Plot for post-application of ticks
plot_data(post_tick_data, 'Days post-application of ticks', 'PCV',
          'PCV vs Days post-application of ticks', 'Days post-application of ticks', 'PCV')
plot_data(post_tick_data, 'Days post-application of ticks', 'PPE',
          'PPE vs Days post-application of ticks', 'Days post-application of ticks', 'PPE')

# Plot for post-inoculation of blood stabilate
plot_data(post_inoculation_data, 'Days post-inoculation of blood stabilate', 'PCV',
          'PCV vs Days post-inoculation of blood stabilate', 'Days post-inoculation of blood stabilate', 'PCV')
plot_data(post_inoculation_data, 'Days post-inoculation of blood stabilate', 'PPE',
          'PPE vs Days post-inoculation of blood stabilate', 'Days post-inoculation of blood stabilate', 'PPE')
```
