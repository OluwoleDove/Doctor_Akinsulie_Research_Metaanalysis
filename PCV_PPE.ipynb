# **IMPORT MODULES**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os

# Mounting Google Drive (if required, skip if already mounted)
from google.colab import drive
drive.mount('/content/drive')
```

# **READ DATA**

```python
# Function to read data from Excel files
def read_data(file_paths):
    data = {}
    for file_name, file_path in file_paths.items():
        df = pd.read_excel(file_path)
        data[file_name] = df
    return data

# Consolidate file paths
file_paths = {
    "c1391bl": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1391bl.xls",
    "c1392bl": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1392bl.xls",
    "c1393bl": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1393bl.xls",
    "c1394bl": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1394bl.xls",
    "c1395bl": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1395bl.xls",
    "c1771": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1771.xls",
    "c1774": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1774.xls",
    "c1775": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1775.xls",
    "c1778": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1778.xls",
    "c1779": "/content/drive/MyDrive/Lekside/repowerbivisualization/c1779.xls",
}

# Read data from all files
data = read_data(file_paths)
```

# **PLOT CHARTS**

```python
# Combine data into a single DataFrame with unified "Days post infection"
combined_data = pd.DataFrame()
for file_name, df in data.items():
    if df['Days post-application of ticks'].notna().any():
        df['Days post infection'] = df['Days post-application of ticks']
    elif df['Days post-inoculation of blood stabilate'].notna().any():
        df['Days post infection'] = df['Days post-inoculation of blood stabilate']
    df['File'] = file_name
    combined_data = pd.concat([combined_data, df], ignore_index=True)

# Filter data to start from day 25
combined_data = combined_data[combined_data['Days post infection'] >= 25]

# Set Seaborn style and context
sns.set(style="darkgrid", context="notebook", palette="muted")

# Plotting function
def plot_data(data, y_col, title, ylabel):
    plt.figure(figsize=(14, 8))
    sns.lineplot(data=data, x='Days post infection', y=y_col, hue='File')
    plt.title(title, fontsize=16, fontweight='bold')
    plt.xlabel('Days post infection', fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    plt.legend(title='Animals', fontsize=12)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    sns.despine(trim=True)
    plt.tight_layout()
    plt.show()

# Plot PCV vs Days post infection
plot_data(combined_data, 'PCV', 'PCV vs Days post infection', 'PCV')

# Plot PPE vs Days post infection
plot_data(combined_data, 'PPE', 'PPE vs Days post infection', 'PPE')
```
